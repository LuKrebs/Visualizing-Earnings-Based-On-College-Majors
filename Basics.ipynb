```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')

recent_grads.iloc[0]
```

```python
raw_data_count = recent_grads.shape[0]

raw_data_count
```

```python
recent_grads = recent_grads.dropna()

cleaned_data_count = recent_grads.shape[0] 

print('Total number of rows in the df, including rows if missing valus')
print(raw_data_count)
print()
print('Number of rows already droped rows with mising values. Using Dropna method')
print(cleaned_data_count)
```

```python
recent_grads.plot(x='Sample_size',
                  y='Median',
                  kind='scatter',
                  title='Sample_size vs. Median',
                  xlim=(-300,5500),
                  ylim=(15000,120000))
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate',
                  kind='scatter',
                  title='Sample_size vs. Unemployment_rate',
                  figsize=(5,5),
                  xlim=(-500,5000),
                  ylim=(-0.02,0.2))
```

```python
recent_grads.plot(x='Full_time',
                  y='Median',
                  kind='scatter',
                  title='Full_time vs. Median',
                  figsize=(5,5),
                  xlim=(-10000,300000),
                  ylim=(-10000, 120000))
```

```python
recent_grads.plot(x='ShareWomen',
                  y='Unemployment_rate',
                  kind='scatter',
                  title='ShareWomen vs. Unemployment_rate',
                  figsize=(5,5),
                  xlim=(-0.1,1.1),
                  ylim=(-0.02, 0.2))
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.scatter(recent_grads['Men'], recent_grads['Median'])
ax1.set_xlim(-15000,200000)
ax1.set_ylim(10000, 120000)
ax1.set_title('Men vs. Median')

ax2.scatter(recent_grads['Women'], recent_grads['Median'])
ax2.set_xlim(-15000,200000)
ax2.set_ylim(10000, 120000)
ax2.set_title('Woman vs. Median')

plt.show()
```
